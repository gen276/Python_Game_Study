```python
# 三目並べ　※コンピュータのアルゴリズムは未導入
# 基本となる線を引く
# クリックすると引いた線の上に〇と×を表示する
# ターン制を導入し、自分が置いた後にコンピューターが0.5秒後にランダムに×を配置する
# 〇か×のどちらかが3つ並んだ場合に勝敗を表示する,両方とも揃う事なく9マス全てが埋まった際には引き分けを表示する
```

```python
import tkinter
import random
import time

masu = [
            [ 0 , 0 , 0 ],
            [ 0 , 0 , 0 ],
            [ 0 , 0 , 0 ]
        ]

# マス全体の印の合計数をカウントする変数を準備する
shirushi = 0
# どちらが勝ったかを判定する変数
kachi = 0
# 勝敗が決定した時のフォントを作成する
FNT = ('Times New Roman' , 60)

# 基本となる線を引く
def masume():
    cvs.delete('all')
    for i in range(1 , 3):
        cvs.create_line(200 * i , 0 , 200 * i , 600 , fill = 'gray' , width = 8)
        cvs.create_line(0 , 200 * i , 600 , 200 * i , fill = 'gray' , width = 8)

    for y in range(3):
        for x in range(3):
            Y = y * 200
            X = x * 200            
            if masu[y][x] == 1 :
                cvs.create_oval(X + 20 , Y + 20 , X + 180 , Y + 180 , outline = 'blue' , width = 12)

            if masu[y][x] == 2 :
                cvs.create_line(X + 20 , Y + 20 , X + 180 , Y + 180 , fill = 'red' , width = 12)
                cvs.create_line(X + 180 , Y + 20 , X + 20 , Y + 180 , fill = 'red' , width = 12)

# ゲームの開始時にはスタートと表示をする
            if shirushi == 0 :
                cvs.create_text(300 , 300 , text = 'スタート!' , fill = 'navy' , font = FNT)
            cvs.update()
                
# 自分がクリックすると〇を表示させる
def click(e):
    global shirushi
    if shirushi == 9 :
        replay()
        return
    
    if shirushi == 1 or shirushi == 3 or shirushi == 5 or shirushi == 7 :
        return
        
# 座標の位置を分かり易く整数で表現する
    mx = int(e.x / 200)
    my = int(e.y / 200)
    if mx > 2 : mx = 2
    if my > 2 : my = 2
    
# クリック時何もなければ〇を表示する
    if masu[my][mx] == 0 :
        masu[my][mx] = 1
        shirushi += 1
        masume()
        time.sleep(0.5)
        hantei()
        syouhai()
        if shirushi < 9 :
            computer()
    else :
            masu[my][mx] = 0
    masume()

# コンピュータの動作を記述
# マスをランダムで選択し、0が入っていた場合にはその地点に2を代入する
def computer():
    global shirushi
    while True :
        x = random.randint(0 , 2)
        y = random.randint(0 , 2)

        if masu[y][x] == 0 :
            masu[y][x] = 2
            shirushi += 1
            masume()
            time.sleep(0.5)
            hantei()
            syouhai()
            break
            
# 判定の動作を記述
# タテ、ヨコ、ナナメのいずれかで同じ値が出た場合に該当するマークの方が勝ち
def hantei():
    global kachi
    kachi = 0    
# タテの場合
    for n in range(1 , 3):
        if masu[0][0] == n and masu[0][1] == n and masu[0][2] == n :
            kachi = n
        if masu[1][0] == n and masu[1][1] == n and masu[1][2] == n :
            kachi = n       
        if masu[2][0] == n and masu[2][1] == n and masu[2][2] == n :
            kachi = n
# ヨコの場合
        if masu[0][0] == n and masu[1][0] == n and masu[2][0] == n :
            kachi = n
        if masu[0][1] == n and masu[1][1] == n and masu[2][1] == n :
            kachi = n
        if masu[0][2] == n and masu[1][2] == n and masu[2][2] == n :
            kachi = n
# ナナメの場合}
        if masu[0][0] == n and masu[1][1] == n and masu[2][2] == n :
            kachi = n
        if masu[2][0] == n and masu[1][1] == n and masu[0][2] == n :
            kachi = n

# 変数の数に応じて勝敗を決定する
def syouhai():
    global shirushi
    if kachi == 1 :
        cvs.create_text(300 , 300 , text = 'あなたの勝ち‼' , font = FNT , fill = 'cyan')
        shirushi = 9
    if kachi == 2 :
        cvs.create_text(300 , 300 , text = 'コンピュータ\nの勝ち‼' , font = FNT , fill = 'gold')
        shirushi = 9
    if kachi == 0 and shirushi == 9 :
        cvs.create_text(300 , 300 , text = '引き分け' , font = FNT , fill = 'lime')

# 全ての内容を空白化して初期値に戻す変数を作成する
def replay():
    global shirushi
    shirushi = 0
    for y in range(3):
        for x in range(3) :
            masu[y][x] = 0
    masume()

root = tkinter.Tk()
root.title('三目並べ')
root.resizable(False , False)
root.bind('<Button>' , click)
cvs = tkinter.Canvas(width = '600' , height = '600' , bg = 'white')
cvs.pack()
masume()
root.mainloop()
```
